```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/Mall_Customers.csv")
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Genre']=le.fit_transform(df[['Genre']])
```

```python
plt.scatter(df['CustomerID'],df['Spending_Score'])
```

```python
import seaborn as sns
sns.pairplot(df)
```

```python
q=df.iloc[:,3:].values
q
```

```python
from sklearn.cluster import AgglomerativeClustering
model = AgglomerativeClustering(
    n_clusters=3,
    metric='euclidean',
    linkage='ward'
)
```

```python
labels=model.fit_predict(df)
labels
```

```python
plt.scatter(q[labels==0,0],q[labels==0,1],color='green',label='cluster1')
plt.scatter(q[labels==1,0],q[labels==1,1],color='blue',label='cluster2')
plt.scatter(q[labels==2,0],q[labels==2,1],color='purple',label='cluster3')
```

```python
from scipy.cluster.hierarchy import dendrogram,linkage
Z = linkage(q, method='ward')
plt.figure(figsize=(8, 5))
dendrogram(Z)
plt.title("Dendrogram")
plt.xlabel("Data Points")
plt.ylabel("Distance")
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=42)
labels = kmeans.fit_predict(q)
labels
```

```python
plt.scatter(q[labels==0,0],q[labels==0,1],color='green',label='cluster1')
plt.scatter(q[labels==1,0],q[labels==1,1],color='blue',label='cluster2')
```
